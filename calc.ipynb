```python
from datetime import datetime
from openpyxl import Workbook
import random

# Function to generate random data for date alternate key, day number of week, day name of week, day number of month,
# week number of year, month number, month name, quarter, semester, and year
def generate_extended_date_data(rows):
    TimeKey= [i for i in range(1, rows + 1)]
    date_alternate_key = [i for i in range(1, rows + 1)]
    day_number_of_week = [random.randint(1, 7) for _ in range(rows)]
    day_name_of_week = [datetime.now().strftime('%A') for _ in range(rows)]  # Getting current day name as an example
    day_number_of_month = [random.randint(1, 31) for _ in range(rows)]
    week_number_of_year = [random.randint(1, 52) for _ in range(rows)]
    day_number_of_year = [random.randint(1, 365) for _ in range(rows)]
    month_number = [random.randint(1, 12) for _ in range(rows)]
    month_name = [datetime.now().strftime('%B') for _ in range(rows)]  # Getting current month name as an example
    quarter = [random.choice(['Q1', 'Q2', 'Q3', 'Q4']) for _ in range(rows)]
    semester = ['Spring' if month <= 6 else 'Fall' for month in month_number]
    year = [datetime.now().year for _ in range(rows)]  # Getting current year as an example

    return list(zip(TimeKey, date_alternate_key, day_number_of_week, day_name_of_week, day_number_of_month,
                    day_number_of_year, week_number_of_year, month_number, month_name, quarter, semester, year))


# Set the number of rows and columns for your Excel file
num_rows = 10
num_columns = 11

# Generate extended date-related data
extended_date_data = generate_extended_date_data(num_rows)

# Combine extended date data with random data
combined_data = [extended_date_data[i] for i in range(num_rows)]

# Create an Excel workbook and select the active worksheet
workbook = Workbook()
sheet = workbook.active

# Define the header for the Excel file
header = ['DateAlternatekey', 'DaynumberOfWeek', 'DayNameOfweek', 'DayNumberOfMonth', 'day_number_of_year',
          'WeekNumberOfYear', 'MonthNumber', 'MonthName', 'Quarter', 'Semester', 'Year']

# Write header to the first row of the worksheet




# Sales data to be added
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
              'October', 'November', 'December'],
    '2011': [10, 7, 12, 10, 19, 19, 7, 22, 12, 14, 6, 25],
    '2012': [12, 7, 16, 10, 19, 21, 21, 19, 12, 12, 9, 21],
    '2013': [11, 10, 13, 11, 19, 21, 16, 27, 11, 14, 10, 19],
    '2014': [12, 12, 15, 10, 17, 20, 15, 20, 11, 7, 11, 17],
    '2015': [11, 9, 20, 13, 19, 21, 18, 22, 13, 9, 11, 13],
    '2016': [13, 11, 22, 14, 11, 20, 12, 23, 13, 11, 9, 11],
    '2017': [20, 16, 27, 21, 21, 25, 23, 33, 22, 15, 14, 35]
}

# Append sales data to the worksheet
for i in range(len(sales_data['Month'])):
    sales_row = [sales_data['Month'][i]] + [sales_data[year][i] for year in sorted(sales_data.keys())]
    sheet.append(sales_row)

# Save the Excel file
excel_filename = 'extended_random_data_with_sales.xlsx'
workbook.save(filename=excel_filename)
```
